```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import confusion_matrix
from word2number import w2n
```

```python
data=pd.read_csv('Automobile_data.csv')
data
```

```python
data['num-of-cylinders']=data['num-of-cylinders'].apply(w2n.word_to_num)
data
```

```python
data.isna().sum()
```

```python
mean_price=data.price.mean()
mean_price
```

```python
data.price=data.price.fillna(mean_price)
data
```

```python
x=data.drop('price', axis=1)
y=data['price']
```

```python
x
```

```python
y
```

```python
le= LabelEncoder()
```

```python
x['company']= le.fit_transform(x['company'])
x['body-style']=le.fit_transform(x['body-style'])
x['engine-type']= le.fit_transform(x['engine-type'])
x
```

```python
data.dtypes
```

```python
x_train, x_test , y_train , y_test= train_test_split(x,y,test_size=0.25, random_state=10)
```

```python
model= RandomForestRegressor(n_estimators=47, random_state=34) 
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
model.score(x_train,y_train)
```

```python
scalar = StandardScaler()
x_train = scalar.fit_transform(x_train)
x_test = scalar.transform(x_test)
```

```python
scaled_model=RandomForestRegressor(max_depth=5)
```

```python
scaled_model.fit(x_train,y_train)
scaled_model.score(x_test, y_test)
```
