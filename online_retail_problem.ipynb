```python
basket_France = data[data['Country']=='France']
basket_France
```

```python
basket_France.isnull().sum()
```

```python
products = basket_France['Description'].unique()
```

```python
basket_France['Description'].value_counts()
```

```python
one_hot = pd.get_dummies(basket_France['Description'])
basket_France = basket_France.join(one_hot)
```

```python
basket_France
```

```python
basket_France=basket_France.drop(['StockCode','Description','Quantity','UnitPrice','CustomerID','Country'],axis=1)
```

```python
basket_France
```

```python
records= basket_France.groupby(['InvoiceNo','InvoiceDate'])[products[:]].apply(sum)
records
```

```python
def get_product_names(p): # p= records dataframe
    for product in products: # products = column names of records
        if p[product]>0:
            p[product]=1
        else:
            p[product] = 0
    return p
records = records.apply(get_product_names, axis=1)
```

```python
records
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
frequent_items = apriori(records, min_support=0.1, use_colnames=True)
rules = association_rules(frequent_items, metric='lift' , min_threshold=1)
```

```python
rules.sort_values(['confidence'] , ascending=False)
```
