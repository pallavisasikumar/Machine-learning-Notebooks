```python
import pandas as pd
from apyori import apriori
```

```python
data=pd.read_csv('Groceries_dataset.csv')
data.head()
```

```python
one_hot=pd.get_dummies(data['itemDescription'])
data=data.join(one_hot)
```

```python
data.head()
```

```python
products= data['itemDescription'].unique()
# proDUCTS
```

```python
data= data.drop('itemDescription', axis=1)
```

```python
data.head()
```

```python
records=data.groupby(['Member_number', 'Date'])[products[:]].apply(sum)
```

```python
records
```

```python
def get_product_names(p):
    #products=['tropical milk','beef','pip fruit']
    for product in products:
        if p[product]>0:
            p[product]= product
    return p
records = records.apply(get_product_names, axis=1)
```

```python
records
```

```python
records.values
```

```python
# removing zeros
transactions = records.values
trans=[]
for i in transactions:
    trans.append(i[i!=0].tolist())
trans
```

```python
len(trans)
```

```python
rules= apriori(trans, min_support=0.0003, min_confidence=0.05, min_lift=3 , min_length=2)
```

```python
association_results=list(rules)
len(association_results)
```
