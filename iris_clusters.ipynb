```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
iris.feature_names
```

```python
x=iris.data
```

```python
sse=[]
for i in range(2,len(iris)):
    km=KMeans(n_clusters=i,random_state=23)
    km.fit(x)
    sse.append(km.inertia_)
```

```python
plt.xlabel('No. of clusters (k)')
plt.ylabel('SSE')
plt.plot(range(2,8),sse[:7])
```

```python
km=KMeans(n_clusters=3, random_state=23)
km.fit(x)
```

```python
km.inertia_
```

```python
clusters= km.fit_predict(x)
clusters
```

```python
km.cluster_centers_
```

```python
df=pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
data_cluster=df.copy()
```

```python
data_cluster['Clusters']= clusters
```

```python
data_cluster
```

```python
data_cluster[data_cluster.Clusters==0]
```

```python
plt.scatter(df['sepal width (cm)'], df['petal width (cm)'], c=clusters)
```
