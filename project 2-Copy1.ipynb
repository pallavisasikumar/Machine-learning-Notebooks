```python
import pandas as pd
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
```

```python
data=pd.read_csv('taylor_swift_spotify.csv')
data
```

```python
data.isnull().sum()
```

```python
data.dtypes
```

```python
data=data.drop(['id','uri','Unnamed: 0','release_date'],axis=1)
```

```python
data
```

```python
x=data.drop('popularity',axis=1)
y=data['popularity']
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le1=LabelEncoder()
```

```python
le2=LabelEncoder()
```

```python
x['name']= le1.fit_transform(x['name'])
x['album']=le2.fit_transform(x['album'])
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_test,x_train,y_test,y_train=train_test_split(x,y,test_size=0.25,random_state=30)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
scaled_model=KNeighborsRegressor()
scaled_model.fit(x_train,y_train)
scaled_model.score(x_test,y_test)

```

```python
knn=KNeighborsRegressor()
knn.fit(x_train,y_train)
knn.score(x_test,y_test)
```

```python
model= RandomForestRegressor(n_estimators=47, random_state=34) 
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
model.score(x_train,y_train)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model=DecisionTreeRegressor()  
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
svm1=SVR()
svm1.fit(x_train,y_train)
svm1.score(x_test,y_test)
```

```python
from sklearn.ensemble import AdaBoostRegressor
```

```python
adaboost= AdaBoostRegressor(n_estimators=50,base_estimator=None, learning_rate=0.3,random_state=1)
```

```python
adaboost.fit(x_train,y_train)
```

```python
adaboost.score(x_test,y_test)
```
