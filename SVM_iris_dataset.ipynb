```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

```python
iris= load_iris()
```

```python
x=iris.data
y=iris.target
x_train , x_test , y_train, y_test = train_test_split(x,y, test_size=0.25, random_state=12)
```

```python
svm=SVC(kernel='linear')
svm.fit(x_train,y_train)
svm.score(x_test,y_test)
```

 #### . gamma = scale : 1/(n_features* X.var())
 #### . gamma = auto:1/(n_features)       

```python
svm_poly = SVC(kernel='poly', degree=4 , C=2.5, gamma='auto') # scale
svm_poly.fit(x_train , y_train)
svm_poly.score(x_test, y_test)
```

```python
svm1= SVC() # rbf kernal (radial basis function)
svm1.fit(x_train , y_train)
svm1.score(x_test, y_test)
```

```python
svm_gauss= SVC(kernel='rbf') # default kernal----- gaussian
svm_gauss.fit(x_train, y_train)
svm_gauss.score(x_test,y_test)
```

```python
svm_sig= SVC(kernel='sigmoid') 
svm_sig.fit(x_train, y_train)
svm_sig.score(x_test,y_test)
```

```python
# machine learning extensions --- for science tasks
# pip install mlxtend
```

```python
from sklearn.decomposition import PCA # principal Component Analysis -- for dimensionally reduction 
from mlxtend.plotting import plot_decision_regions
```

```python
pca = PCA(n_components=2)
x_train_pca = pca.fit_transform(x_train)
x_test_pca = pca.transform(x_test)
```

```python
svm_pca =SVC()
svm_pca.fit(x_train_pca,y_train)
plot_decision_regions(x_train_pca, y_train, clf=svm_pca)
plt.figure()
plot_decision_regions(x_test_pca,y_test, clf=svm_pca)
svm_pca.score(x_test_pca,y_test)
```

```python
svm_pca =SVC(kernel='poly')
svm_pca.fit(x_train_pca,y_train)
plot_decision_regions(x_train_pca, y_train, clf=svm_pca)
plt.figure()
plot_decision_regions(x_test_pca,y_test, clf=svm_pca)
svm_pca.score(x_test_pca,y_test)
```

```python
svm_pca =SVC(kernel='poly', C=0.25)
svm_pca.fit(x_train_pca,y_train)
plot_decision_regions(x_train_pca, y_train, clf=svm_pca)
plt.figure()
plot_decision_regions(x_test_pca,y_test, clf=svm_pca)
svm_pca.score(x_test_pca,y_test)
```

```python
svm_pca =SVC(kernel='poly', C=0.25,degree=5)
svm_pca.fit(x_train_pca,y_train)
plot_decision_regions(x_train_pca, y_train, clf=svm_pca)
plt.figure()
plot_decision_regions(x_test_pca,y_test, clf=svm_pca)
svm_pca.score(x_test_pca,y_test)
```

```python
svm_pca =SVC(kernel='poly', C=0.25,degree=5,gamma='auto')
svm_pca.fit(x_train_pca,y_train)
plot_decision_regions(x_train_pca, y_train, clf=svm_pca)
plt.figure()
plot_decision_regions(x_test_pca,y_test, clf=svm_pca)
svm_pca.score(x_test_pca,y_test)
```

```python
svm_pca =SVC(kernel='linear')
svm_pca.fit(x_train_pca,y_train)
plot_decision_regions(x_train_pca, y_train, clf=svm_pca)
plt.figure()
plot_decision_regions(x_test_pca,y_test, clf=svm_pca)
svm_pca.score(x_test_pca,y_test)
```

```python
svm_pca =SVC(kernel='sigmoid')
svm_pca.fit(x_train_pca,y_train)
plot_decision_regions(x_train_pca, y_train, clf=svm_pca)
plt.figure()
plot_decision_regions(x_test_pca,y_test, clf=svm_pca)
svm_pca.score(x_test_pca,y_test)
```
