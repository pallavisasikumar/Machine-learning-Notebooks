```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data=pd.read_csv('canada_per_capita_income.csv')
data
```

```python
plt.xlabel('Year')
plt.ylabel('Per capita income')
plt.scatter(data.year, data['per capita income (US$)'], color='blue' , marker='*')
```

```python
x=data.drop('per capita income (US$)', axis=1)  
y= data['per capita income (US$)']  
```

```python
x
```

```python
y
```

```python
model=LinearRegression()
model.fit(x,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred=model.predict(x)
```

```python
pd.DataFrame({'True y values:':y , 'Predicted y values':y_pred})
```

```python
plt.scatter(x,y)
plt.xlabel('Year')
plt.ylabel('Per capita income')
plt.plot(data['year'], y_pred, color='red')  
```

```python
model.predict([[2022]])
```
