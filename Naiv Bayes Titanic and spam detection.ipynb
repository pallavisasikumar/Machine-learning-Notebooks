```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB # for hamdling descrete values
```

```python
data=pd.read_csv('spam.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
data['spam']= data.Category.apply(lambda x: 1 if x=='spam' else 0)
```

```python
data.head()
```

```python
data.drop('Category', axis=1)
data.head()
```

```python
x= data.Message
y= data.spam
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25, random_state=37)
```

```python
x_train.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer()  # its a class ..so object is created here
x_train_counter = cv.fit_transform(x_train.values)
```

```python
cv.get_feature_names()
```

```python
pd.DataFrame(x_train_counter.toarray(), columns= cv.get_feature_names())
```

```python
x_train_counter.shape
```

```python
model=MultinomialNB()
```

```python
model.fit(x_train_counter, y_train)
```

```python
model.score(cv.transform(x_test),y_test)
```

```python
# predictions 

emails=['Can we get together to watch football game tommorow?',
        'upto 20% discount on parking, exclusive offer just for you']
```

```python
email_counter = cv.transform(emails)
model.predict(email_counter)
```
